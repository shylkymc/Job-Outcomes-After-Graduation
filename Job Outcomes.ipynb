### Exploring with Scatter Plots

I will visualize the "Sample_size" and "Median" columns.

```python
recent_grads.plot(x="Sample_size", y="Median", kind="scatter", title= "Median vs. Sample_size")
```

Let's take a look to the employment situation.

```python
recent_grads.plot(x='Sample_size', y='Employed', kind='scatter', title='Employed vs. Sample_size', figsize=(5,10))
```

```python
recent_grads.plot(x="Unemployment_rate", y="Sample_size",  kind="scatter", title= 'Unemployed vs. Sample_size', figsize=(10,5))
```

I will check the full time and part time employments situation.

```python
recent_grads.plot( x="Median",y="Full_time", kind="scatter")
```

```python
recent_grads.plot( x="Median",y="Part_time", kind="scatter")
```

I will check the women's and men's situation.

```python
recent_grads.plot(x="ShareWomen", y="Unemployment_rate", kind="scatter")
```

```python
recent_grads.plot( x="Median", y="Women", kind="scatter", title= "Employed Women")
```
