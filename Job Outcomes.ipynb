```python
columns= ['Sample_size', 'Median','Employed', 'Full_time', 'ShareWomen', 'Unemployment_Rate', 'Men','Women']
fig= plt.figure(figsize=(5,12))
for r in range(1,5):
    ax= fig.add_subplot(2,2,r)
    ax= recent_grads[columns[r]].plot(kind='hist', rot=40)
```
